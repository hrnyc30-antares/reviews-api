```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('reviews.csv')
df2 = pd.read_csv('reviews_photos_original.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
merged = df1.merge(df2, left_on="id", right_on="review_id", how="right").fillna("")
```

```python
merged.head()
```

```python
merged.drop(['id_x','product_id', 'rating', 'summary', 'body', 'recommend', 'reported', 'reviewer_name', 'reviewer_email', 'response', 'helpfulness'], inplace=True, axis=1)
```

```python
merged.head()
```

```python
merged._id=merged._id.str[9:-1]
```

```python
merged.head()
```

```python
merged = merged.rename(columns={"review_id": "old_review_id", "_id":"review_id", "id_y":"old_id"})
merged.head()
```

```python
merged.to_csv("reviews_photos.csv", index=False)
```
